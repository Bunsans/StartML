```python
unique_playstore = playstore.drop_duplicates(subset = ['App'], ignore_index=True)
```

```python
unique_playstore_ = unique_playstore.copy()
unique_playstore_['price'] = unique_playstore_['price'].str.replace('$', '', regex=True) 

unique_playstore_['price'] = unique_playstore_['price'].astype({'price': float})
```

```python
unique_playstore_[unique_playstore_['price'] != '0']
```

```python
tmp = unique_playstore_.copy()
res = pd.pivot_table(tmp,
                    index= ['category', 'type'],
                    values=['rating', 'reviews', 'price'],
                    aggfunc='mean')
res = res.rename(columns={'price': 'mean_price' , 'rating': 'mean_rating', 'reviews': 'mean_reviews'})
res['mean_price'] = res['mean_price'].round(2)
res['mean_rating'] = res['mean_rating'].round(1)
res['mean_reviews'] = res['mean_reviews'].round(2)
```

```python
res.to_csv('result.csv', sep=',')
```
