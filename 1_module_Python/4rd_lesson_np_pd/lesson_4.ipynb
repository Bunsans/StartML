````python
vec1 = list(range(int(1e5)))
vec2 = list(range(int(1e5)))
````

````python
def dot_(vec1, vec2):
    res = 0
    for i in range(len(vec1)):
        res += vec1[i] * vec2[i]
    return res
    
````

````python
%%timeit
res = dot_(vec1,vec2)
print("{:.2e}".format(res))
````
